```python
#south side
WEATHER_SOUTH_LAT='53.335996'
WEATHER_SOUTH_LON='-6.262891'
#north side
WEATHER_NORTH_LAT=''
WEATHER_NORTH_LAT=''
# https://api.openweathermap.org/data/2.5/weather?lat=53.343026&lon=-6.27717&appid=74571a8a54ecc764ebed7d0dc725ebe2
REQUEST='https://api.openweathermap.org/data/2.5/weather/'
# https://api.jcdecaux.com/vls/v1/stations?contract=dublin&apiKey=06af51839f293cccf02a9ba5f4204bb094506a95
WAPIKEY=''

''' WHAT DOES THIS DO EXACTLY?'''
def write_to_file(text):
    date_underscores = "{:%Y_%m_%d_%H_%M_%S}".format(datetime.datetime.now())
    with open(r"data\weather_{}".format(date_underscores).replace(" ", "_"), "w") as f:
        f.write(text)
        
def weather_to_db(text):
        weather = json.loads(text)
        print(type(weather), len(weather))
        for weat in weather: 
            print(weat)
            vals = (weat.get("coord").get("lon"),
                    weat.get("coord").get("lat"),
                    weat.get("weather").get("id"),
                    weat.get("weather").get("main"),
                    weat.get("weather").get("description"),
                    weat.get("weather").get("icon"),
                    weat.get("base"),
                    weat.get("main").get("temp"),
                    weat.get("main").get("feels_like"),
                    weat.get("main").get("temp_min"),
                    weat.get("main").get("temp_max"),
                    weat.get("main").get("pressure"),
                    weat.get("main").get("humidity"),
                    weat.get("visibility"),
                    weat.get("wind").get("speed"),
                    weat.get("wind").get("deg"),
                    weat.get("rain").get("1h"),
                    weat.get("clouds").get("all"),
                    weat.get("dt"),
                    weat.get("sys").get("type"),
                    weat.get("sys").get("id"),
                    weat.get("sys").get("country"),
                    weat.get("sys").get("sunrise"),
                    weat.get("sys").get("sunset"),
                    weat.get("timezone"),
                    weat.get("id"),
                    weat.get("name"),
                    weat.get("cod"),
            engine.execute("INSERT INTO `dublin_bikes`.`open_weather_south` values (%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s)", vals)
        return

def main():
    print(os.path)
    while True:
        try:
            now = datetime.datetime.now()
            r = requests.get(REQUEST, params={"appid": WAPIKEY, "lon": WEATHER_SOUTH_LAT, "lat":WEATHER_SOUTH_LON}
            print(r, now)
            write_to_file(r.text)
            weather_to_db(r.text)
            time.sleep(5*60)
        except:
            print(traceback.format_exc())
             #if engine is None:
                #pass
        
    return 
if __name__== "__main__":
    main()                 
            
```
